```python
from statsmodels.tsa.seasonal import seasonal_decompose
rv_seasonal = seasonal_decompose(rv['total'])
rv_seasonal.plot();
```

```python
train_rv = rv.iloc[0:450]    # 1st 450 in train
test_rv = rv.iloc[450:479]   # last few records in the test
```

```python
from pmdarima import auto_arima

auto_arima(rv['total'], seasonal = True, m=7)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model1 = SARIMAX(train_rv['total'], order=(1, 0, 0), seasonal_order=(1, 0, 1, 7))
results = model1.fit()
```

```python
pred_rv = results.predict(start=450, end = 477)
pred_rv
```

```python
print(test_rv.shape)
print(pred_rv.shape)
```

```python
test_rv[test_rv['holiday']==1].index
```

```python
plt.figure(figsize= (12,6))
plt.plot(test_rv.total, marker='*', color='g', label='Actual_test')
plt.plot(pred_rv, color = 'r', label = 'Pred_test', marker='*')
plt.legend()
plt.axvline(x='2017-04-14' , color='y')
plt.axvline(x='2017-04-16' , color='y')
plt.axvline(x='2017-04-17' , color='y')
```

```python
error = test_rv.total - pred_rv
MSE = np.mean(np.square(error))
print('MSE in SARIMA: ', MSE)
RMSE = np.sqrt(MSE)
print('RMSE in SARIMA:', RMSE)
```

#### Till this point only SARIMA is used and not SARIMAX i.e value of 'X' is not passed yet

```python
from pmdarima import auto_arima

auto_arima(rv['total'], seasonal = True, m=7, exogenous = rv[['holiday']])

# exogenous is always passed as dataframe format because more than 1 columns can be taken as X
# you can pass multiple exogenous values in the model
```

```python
model2 = SARIMAX(train_rv['total'],
                 order=(1, 0, 0),
                 seasonal_order=(1, 0, 1, 7),
                 exog = train_rv[['holiday']])
results1 = model2.fit()
```

```python
pred_rv_exos = results1.predict(start=450, end=477, exog = test_rv[['holiday']])
pred_rv_exos
```

```python
error = test_rv.total - pred_rv_exos
MSE = np.mean(np.square(error))
print('MSE in SARIMAX : ', MSE)
RMSE = np.sqrt(MSE)
print('RMSE in SARIMAX:', RMSE)
```

```python
plt.figure(figsize= (12,6))
plt.plot(test_rv.total, marker='*', color='g', label='Actual_test')
plt.plot(pred_rv_exos, color = 'r', label = 'Pred_test', marker='*')
plt.legend()
plt.axvline(x='2017-04-14' , color='y')
plt.axvline(x='2017-04-16' , color='y')
plt.axvline(x='2017-04-17' , color='y')
```
